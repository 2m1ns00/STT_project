(1) unexpected gender : 0 -> 기존 AUDIO_INFO 는 AUDIO_INFO_origin -> 민수가 만든 AUDIO_INFO는 AUDIO_INFO로 수정

(2) Unexpected gender: 'f f' -> TH 발화자들이 train/ test에서 겹치는 문제로 인식 -> train/test 겹치지 않도록함

(3) 004 TH0128 f 까지 완료 / Invalid line in utt2spk file: CN0001_004 -> 
  1. utt2spk, text, wav.scp는 직접 추가(x), 
  2. 모델돌아가면서 utt2spk에 CN0001_004 형식으로 새로 작성되는 것 발견
  3. 폴더에 flac이랑 txt 제외한 파일들 제거, trans.txt 파일에 빈줄과 중복 내용확인

(4) 음성모델까지 잘 돌아감
