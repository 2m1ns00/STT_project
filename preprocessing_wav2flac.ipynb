```python
import os
from pathlib import Path

fileDir = r"/opt/bin/daehan/data/train"
fileExt1 = r".wav"
fileExt2 = r".flac"
wav_lst = set([f[:-4] for f in os.listdir(fileDir) if f.endswith(fileExt1)])
flac_lst = set([f[:-9] for f in os.listdir(fileDir) if f.endswith(fileExt2)])

print(len(wav_lst),len(flac_lst))
print("찾았다",wav_lst.difference(flac_lst))
```
